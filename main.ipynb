```python
#plik odpowiedzialny w większości za UI programu

%run logic.ipynb
import random
import tkinter as tk 
from tkinter import *
frame = tk.Tk() 
frame.title("Spam filter - Piotr Katolik 162100") 
frame.geometry('400x350') 
#funkcja odpowiedzialna za sprawdzenie czy plik jest spamem
def spam_check(): 
    inp = inputtxt.get(1.0, "end-1c") 
    inp_conv=transform_text(inp) # type: ignore
    inp_conv_test=tfidf.transform([inp_conv]) # type: ignore
    result=model.predict(inp_conv_test)[0] # type: ignore
    if result == 1:
        result_text="Spam"
        lbl.config(fg='#F00')
    else:
        result_text="not Spam"
        lbl.config(fg='#0F0')
    lbl.config(text = ""+result_text)
    
#funkcja generowania spamu
def spam_gen():
    inputtxt.delete('1.0', END)
    s=test_data.loc[test_data['CATEGORY'] == 1] # type: ignore
    val = s['MESSAGE'].values[random.randint(0, 5)]
    inputtxt.insert(END,val) 
    
#funkcja generowania hamu
def ham_gen():
    inputtxt.delete('1.0', END)
    s=test_data.loc[test_data['CATEGORY'] == 0] # type: ignore
    val = s['MESSAGE'].values[random.randint(0, 5)]
    inputtxt.insert(END,val) 
        
#utworzenie etykietu pokazującego czy tekst jest spamem
lbl = tk.Label(frame, text = "") 
lbl.grid(row = 0, column = 4) 
def enterfunction(event=' '):
    spam_check()
    
#zbindowanie przycisku enter do pokazania rezultatu
frame.bind('<Return>', enterfunction)

#pole tekstowe do wczytania/wpisania danych
inputtxt = tk.Text(frame,height = 10, width = 30)
inputtxt.grid(row = 0, column = 0,columnspan=3)

#utworzenie przycisku generate spam
CheckButton = tk.Button(frame, text = "Generate Spam", command = spam_gen) 
CheckButton.grid(row = 2, column = 0)

#utworzenie przycisku spam check
CheckButton = tk.Button(frame, text = "Spam Check", command = spam_check) 
CheckButton.grid(row = 2, column = 1)

#utworzenie przycisku generate ham
CheckButton = tk.Button(frame, text = "Generate Ham", command = ham_gen) 
CheckButton.grid(row = 2, column = 2)

#utworzenie pola tekstowego z ocenami modelu
text = tk.Text(frame,width=20)
text.insert(1.0,(Scores_String)) # type: ignore
text.grid(row = 3, column = 1,sticky=S)
text.config(state='disabled')

frame.mainloop() 
```
